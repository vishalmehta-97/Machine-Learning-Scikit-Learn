```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

# “A large synthetic student performance dataset was used to train and evaluate machine learning regression models.”

```python
data=pd.read_csv("./Data/student_performance_dataset.csv")
data.head()
```

```python
X=data[["Study_Hours_per_Week"]]
y=data["Final_Score"]
```

```python
model=LinearRegression()
model.fit(X,y)
predicted_scores=model.predict(X)
```

```python
mae=mean_absolute_error(y,predicted_scores)
mse=mean_squared_error(y,predicted_scores)
rmse=np.sqrt(mse)
R2_score=r2_score(y,predicted_scores)
```

```python
print(round(mae,2))
print(round(mse,2))
print(round(rmse,2))
print(round(R2_score,4))
```

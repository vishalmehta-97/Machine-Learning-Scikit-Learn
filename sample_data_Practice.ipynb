# ENCODING CATOGORIAL VALUE

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()     #(LabelEncoding)
```

```python
data["Gender_encoded"]=le.fit_transform(data["Gender"])
data["Passed_encoded"]=le.fit_transform(data["Passed"])
```

```python
data   ## With this we have encoded the Gender Column (LabelEncoding)
```

```python
## One-Hot Encoding
data_encoded=pd.get_dummies(data,columns=["City"], dtype=int)
```

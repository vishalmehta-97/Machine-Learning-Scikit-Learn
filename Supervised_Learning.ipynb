## LINEAR REGRESSION
1. Finds Relationship Btw Input & Output (Past Data) 
2. Forms a Straight Line
3. Predicts Numeric Values

```python
from sklearn.linear_model import LinearRegression
X=[[1],[2],[3],[4],[5]]
y=[40,50,65,75,90]
model=LinearRegression()
```

```python
model.fit(X,y)

study_hours=float(input("Total No. Of Hours You Studied = "))

predicted_marks=model.predict([[study_hours]])

f"Based on your study hours {study_hours} you can get {predicted_marks} marks"
```

## CLASSIFICATION

# 1. LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
X=[[1],[2],[3],[4],[5]]
y=[0,0,1,1,1]  ## 0--> Fail, 1--> Pass
```

```python
model.fit(X,y)
study_hours=float(input("Total No. Of Hours You Studied = "))
result=model.predict([[study_hours]])[0]
if result==1:
    print(f"Based on the no of hours you studied ! You are Likely to Pass")
else:
    print(f"Based on the no of hours you studied ! You are Likely to Fail")
```

# 2. K-NEAREST NEIGHBORS (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=3)
```

```python
X=[
    [180 ,7],
    [200 ,7.5],
    [250 ,8],
    [300 ,8.5],
    [330 ,9],
    [360 ,9.5]
]
y=[0,0,0,1,1,1]    ## 0--> Apple & 1--> Orange
```

```python
model.fit(X,y)

weight=float(input("Enter the weight in grams: "))
size=float(input("Enter the weight in cm: "))

prediction=model.predict([[weight,size]])[0]   ## WE USE [0] TO GET THE FIRST ELEMEMT FROM THE ARRAY
 
if prediction==0:
    print("This is likely an Apple")
else:
    print("This is likely an Orange")
```

# DECISION TREE   

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
```

```python
X=[
    [7,2],    # apple
    [8,3],    # apple
    [9,8],    # Orange
    [10,9]   # Orange
]

y=[0,0,1,1]  # 0--> Apple,1--> Orange
```

```python
model.fit(X,y)

size=float(input("Enter the weight in cm: "))
shade=float(input("Enter the colorshade (1-10) : "))

result= model.predict([[size, shade]])[0]

if result==0:
    print("This is likely an Apple")
else:
    print("This is likely an Orange")
```

```python
import pandas as pd
```

```python
data=pd.read_csv("./Data/practice_data.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data["Salary"]=data["Salary"].fillna(data["Salary"].mean())
data["Experience"]=data["Experience"].fillna(data["Experience"].mean())
data["Age"]=data["Age"].fillna(data["Age"].mean())
```

```python
data
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
data["GenderEncoded"]=le.fit_transform(data["Gender"])
data["PassedEncoded"]=le.fit_transform(data["Passed"])
```

```python
data
```

```python
Encoded_data=pd.get_dummies(data,columns=["City"],dtype=int)
```

```python
pre_data=Encoded_data[["Name","GenderEncoded","Age","Salary","Experience","PassedEncoded","City_Bangalore","City_Chennai","City_Delhi","City_Mumbai"]]
```

```python
pre_data
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
Standard_scaled=scaler.fit_transform(pre_data[["Age","Salary","Experience"]])
```

```python
pd.DataFrame(Standard_scaled,columns=["Age","Salary","Experience"])
```

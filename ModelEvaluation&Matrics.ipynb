# CLASSIFICATION METRICS

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# True answers (what actually happened)
y_true=[1,0,1,1,0,1,0]

#Model's prediction (what is guessed)
y_pred=[1,0,1,0,0,1,1]
```

## Evaluation

```python
"Accuracy:", accuracy_score(y_true, y_pred)
```

```python
"Precision:", precision_score(y_true, y_pred)
```

```python
"Recall:", recall_score(y_true, y_pred)
```

```python
"F1 Score:", f1_score(y_true, y_pred)
```

# REGRESSION MATRICS

## MAE (MEAN ABSOLUTE ERROR)
## MSE (MEAN SQUARED ERROR)
## RMSE (ROOT MEAN SQUARED ERROR)

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
```

```python
## real scores
real_scores=[90,60,80,100]

## Model guess
predicted_scores=[85,70,70,95]
```

```python
mae=mean_absolute_error(real_scores, predicted_scores)
mse=mean_squared_error(real_scores, predicted_scores)
```

```python
"MAE:", mae
```

```python
"MSE:", mse
```

```python
rmse=np.sqrt(mse)  
print("RMSE",rmse)
```
